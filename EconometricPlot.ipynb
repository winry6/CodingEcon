```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Wage intervals and corresponding frequencies
wage_intervals = [10, 20, 30, 40]  # Upper bounds of intervals
frequencies = [0.1, 0.4, 0.3, 0.2]  # Frequencies for the intervals
intervals = [(0, 10), (10, 20), (20, 30), (30, 40)]
```

```python
# Plot the discrete distribution with mass at the right-most points
plt.figure(figsize=(8, 6))
plt.stem(wage_intervals, frequencies, basefmt=" ")
plt.xlabel('Wage')
plt.ylabel('Frequency')
plt.title('Discrete Distribution (Mass at Right-most Points)')
plt.grid(True)
plt.show()
```

```python
# Left-most points
left_most_points = [0, 10, 20, 30]

# Plot the discrete distribution with mass at the left-most points
plt.figure(figsize=(8, 6))
plt.stem(left_most_points, frequencies, basefmt=" ")
plt.xlabel('Wage')
plt.ylabel('Frequency')
plt.title('Discrete Distribution (Mass at Left-most Points)')
plt.grid(True)
plt.show()
```

```python
# Define the uniform density function over each interval
def uniform_density(x):
    if 0 <= x <= 10:
        return 0.1 / 10
    elif 10 < x <= 20:
        return 0.4 / 10
    elif 20 < x <= 30:
        return 0.3 / 10
    elif 30 < x <= 40:
        return 0.2 / 10
    else:
        return 0

# Generate x values for the plot
x_vals = np.linspace(0, 40, 400)
y_vals = [uniform_density(x) for x in x_vals]

# Plot the uniform distribution
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals)
plt.xlabel('Wage')
plt.ylabel('Density')
plt.title('Uniform Distribution on Each Interval')
plt.grid(True)
plt.show()
```

```python
# Define the cumulative distribution function (CDF)
def distribution_function(x):
    if x < 0:
        return 0
    elif 0 <= x <= 10:
        return 0.1 * (x / 10)
    elif 10 < x <= 20:
        return 0.1 + 0.4 * ((x - 10) / 10)
    elif 20 < x <= 30:
        return 0.5 + 0.3 * ((x - 20) / 10)
    elif 30 < x <= 40:
        return 0.8 + 0.2 * ((x - 30) / 10)
    else:
        return 1
```

```python
# Generate x values for plotting
x_vals = np.linspace(0, 40, 400)
# Evaluate the distribution function (CDF) for all x values
y_cdf = [distribution_function(x) for x in x_vals]

# Plot the distribution function (CDF)
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_cdf, label="CDF", color='orange')
plt.xlabel('Wage')
plt.ylabel('Cumulative Probability')
plt.title('Cumulative Distribution Function (CDF)')
plt.grid(True)
plt.show()
```
